```python
from keras.datasets import mnist
(train_images, train_labels),(test_image, test_labels)=mnist.load_data()
train_images.shape
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = test_image
Y = test_labels
```

```python
digit = train_images[22]
plt.imshow(digit,cmap=plt.cm.binary)
plt.show
print('số này là:',train_labels[22])
```

```python
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.tight_layout()
    plt.imshow(train_images[i],cmap='gray',interpolation='none')
    plt.title('số:{}'.format(train_labels[i]))
    plt.xticks([])
    plt.yticks([])
```

```python
from keras.models import Sequential
from keras.layers import Dense
model = Sequential()
model.add(Dense(512,activation='relu',input_shape = (784,)))
```

```python
model.add(Dense(10,activation='softmax'))
model.summary()
```

```python
# chuyển dữ liệu về từ 0-255 về 0-1
train_images = train_images.reshape(60000,28*28)
train_images = train_images.astype('float32')/255
test_images = test_image.reshape(10000,28*28)
test_images = test_image.astype('float32')/255
```

```python
from keras .utils import to_categorical
train_labels = to_categorical(train_labels,10)
test_labels = to_categorical(test_labels,10)
```

```python
model.compile(optimizer='rmsprop',loss='categorical_crossentropy',metrics=['accuracy'])
model.fit(train_images, train_labels,epochs=10,batch_size=128)
test_loss,test_acc = model.evaluate(test_image,test_labels)
print('Mất mát:',test_loss)
print('Độ chính xác:',test_acc)            
```

```python
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import cifar10
from keras.utils.np_utils import to_categorical   
(X_train, Y_train), (X_test, Y_test) = cifar10.load_data()
```

```python
X = X_test
Y = Y_test
```

```python
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.tight_layout()
    plt.imshow(X_train[i],cmap='gray',interpolation='none')
    plt.title('số:{}'.format(Y_train[i]))
    plt.xticks([])
    plt.yticks([])
```

```python
from keras.models import Sequential
from keras.layers import Dense
model = Sequential()
model.add(Dense(1000,activation='relu',input_shape = (3072,)))
```

```python
model.add(Dense(10,activation='softmax'))
model.summary()
```

```python
X_train = X_train.reshape(50000,32*32*3)
X_train = X_train.astype('float32')/255
X_test = X_test.reshape(10000,32*32*3)
X_test = X_test.astype('float32')/255
```

```python
from keras .utils import to_categorical
Y_train = to_categorical(Y_train,10)
Y_test = to_categorical(Y_test,10)
```

```python
model.compile(optimizer='rmsprop',loss='categorical_crossentropy',metrics=['accuracy'])
model.fit(X_train, Y_train,epochs=10,batch_size=128)
test_loss,test_acc = model.evaluate(X_test,Y_test)
print('Mất mát:',test_loss)
print('Độ chính xác:',test_acc)
```
